**About The Program**

This Program takes few inputs like command, component name, wattage of component, working hours in a day.
There are two types of commands. when you are asked to enter the command. Please enter "continue" to calculate
unit consumption details for a particular component and when you finally calculated unit consumption tasks for 
all the components you would like to, please enter "generate" to generate the final bill by listing all the 
components along with their unit consumption details. Unit price is different for every country/region. multiply number of final units with your area unit price to get your electricity bill price.

I have ran the program for sample purpose.

```python
print("Python program to calculate electricity units consumed by various components")
print("please follow the instructions below and enter data accordingly")
print("while program asks you to enter command please enter 'continue' to calculate\n unit consumption details and type 'generate' to\n generate final bill")
    
bill_list = {}


def user_int():
    print("enter command: ")
    command = input("\ntype command here: ")
    if command == "generate":
        print("Summary")
        for key, value in bill_list.items():
            print(key," : ",value)
            
            
        print("total unit consumption is: ", sum(bill_list.values()))
        print("Thankyou for using this program")

    else:
        comp = input("enter component name: \n")
        wattage = float(input("\nenter wattage of the component: \n"))
        hours = float(input("\nenter number of working hours in a day: \n"))
        days = float(30)
        bill = float((wattage/1000)*hours*days)
        bill_list.update({comp : bill})
        print("\nthe electricity units consumption for", comp,"for one month is", bill)
        print("\ntype 'continue' to calculate bill for another component")
        print("type 'generate' to calculate total bill")
        user_int()

user_int()
     
    
        



    
```
